```python
result = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
print(result)
```

```python
print(result[0])
```

```python
df = pd.DataFrame(result[0])
df.head(12)
```

```python
df = df.rename(columns={'Postal Code': 'PostalCode', 'Neighbourhood': 'Neighborhood'})
df.head(10)
```

```python
df.drop(df.index[df['Borough'] == 'Not assigned'], inplace = True)
df = df.reset_index(drop=True)
df.head(12)
```

```python
dup = df[df['Postal Code'].duplicated()]
dup
```

```python
df_shape = df.shape
"The data has {} rows, and {} colums".format(*df_shape)
```

```python
geospatial = pd.read_csv('http://cocl.us/Geospatial_data')
geospatial
```

```python
geospatial = geospatial.rename(columns={'Postal Code': 'PostalCode'})
geospatial
```

```python
df = pd.merge(left=df, right=geospatial, how='inner', on='PostalCode')
df
```

```python
address = 'Toronto, ON' # creating a variable and assigning address parameters to it

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# creating a map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

neighborhood = df['Neighborhood']
borough = df['Borough']
lat = df['Latitude']
lng = df['Longitude']

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
